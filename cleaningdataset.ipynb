```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('https://storage.googleapis.com/codehub-data/1-A-2-4-sales1.csv')
df2 = pd.read_csv('https://storage.googleapis.com/codehub-data/1-A-2-4-sales2.csv')
```

```python
df=pd.merge(df1, df2, on = "Order ID", how = "outer")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
result = df[df['Region'].isnull()]
result
```

```python
df['Region'] = df['Region'].fillna("unknown")
df['Country'] = df['Country'].fillna("unknown")
```

```python
df['Order Priority'] = df['Order Priority'].fillna(df['Order Priority'].mode()[0])
#     df['Order Priority'] accesses the "Order Priority" column in your DataFrame.
#     .fillna() is a method in pandas that fills missing (NaN) values with the specified value.
#     df['Order Priority'].mode()[0] calculates the mode of the "Order Priority" column and selects the first (most frequent) value.
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'].isnull()[900:930]
```

```python
summary = df.groupby(["Item Type"],as_index=True)["Unit Cost"].agg(['count','mean','std'])
summary   
```

```python
df['Total Profit'] = df['Total Revenue']-df['Total Cost']
```

```python
df['Total Profit'].isnull().sum()
```

```python
df['Total Profit']= df['Total Profit'].fillna(0)
```
